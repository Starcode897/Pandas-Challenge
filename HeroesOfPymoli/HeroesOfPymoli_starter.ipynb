```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(780)
```

## Player Count

* Display the total number of players


```python
print(purchase_data["SN"].nunique())
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item Name"].nunique()
average_price = purchase_data["Price"].mean()
number_purchases = purchase_data["SN"].count()
total_revenue = purchase_data["Price"].sum()

Purchase_Analysis = [{"Number of Unique Items": unique_items, "Average Price": average_price,
               "Number of Purchases": number_purchases, "Total Revenue": total_revenue}]
Purchase_df = pd.DataFrame(Purchase_Analysis)
Purchase_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Identifies the amount of players in respect to their gender
male_players = (purchase_data["Gender"] == 'Male').sum()
female_players = (purchase_data["Gender"] == 'Female').sum()
nd_players = (purchase_data["Gender"] == 'Other / Non-Disclosed').sum()

#Identifies the percent
percent_male = ((male_players / purchase_data["Gender"].count()) * 100).round(decimals=1)
percent_female = ((female_players / purchase_data["Gender"].count()) * 100).round(decimals=1)
percent_nd = ((nd_players / purchase_data["Gender"].count()) * 100).round(decimals=1)

print(f"Percent/Count Male: {percent_male}")
print(f"Percent/Count Female: {percent_female}")
print(f"Percent/Count Non-Disclosed: {percent_nd}")
purchase_data.groupby("Gender").nunique()["SN"]
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by gender

gender_group = purchase_data.groupby(["Gender"])

#Counts unique users by gender
gender_count = gender_group.nunique()["SN"]

#The rest
average_price = gender_group["Price"].mean().round(2)

purchase_count = gender_group["Price"].count()

total_value = gender_group["Price"].sum()

per_person = (total_value / gender_count).round(2)

gender_table = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_price,
               "Total Purchase Value": total_value, "Total Average Price per Person": per_person})

gender_table


#Observable trend 1

#Due to the lower amount of female players, the average purchase price is higher. 
#this is due to a lower distribution of values. The same could be seen for Other/Non-Disclosed.
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python


age_bins = [0, 10, 20, 30, 40, 50]
names = ["<10", "11-20", "21-30", "31-40", "41-49"]

#Uses the aggregate function to split the count of players into each bin.
group = pd.cut(purchase_data["Age"], age_bins, labels=names, include_lowest=True)
age_demo_df = purchase_data.groupby(group)["Age"].agg(["count"]).rename(columns={'count': 'Total Count'})

#Adds another column for percent of players in each demographic.
age_demo_df["Percentage of Players"] = ((age_demo_df["Total Count"] / 780) * 100).round(decimals=2)
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python


age_bins2 = [0, 10, 20, 30, 40, 50]
names2 = ["<10", "11-20", "21-30", "31-40", "41-49"]

#Uses the bins to categorize the data
group2 = pd.cut(purchase_data["Age"], age_bins2, labels=names2, include_lowest=True)
purchase_ana = purchase_data.groupby(group2) 



gender_count2 = purchase_ana.nunique()["SN"]

average_price2 = purchase_ana["Price"].mean().round(2)

purchase_count2 = purchase_ana["Price"].count()

total_value2 = purchase_ana["Price"].sum()

per_person2 = (total_value2 / gender_count2).round(2)
    
purchase_table = pd.DataFrame({"Purchase Count": purchase_count2, "Average Purchase Price": average_price2,
                               "Total Purchase Value": total_value2, "Total Average Price per Person": per_person2})
purchase_table


    


#Observable trend 2

#Most of the spenders exist within the 21-30 age bracket, and as a result they're spending the most money. 
#Makes sense since those people would probably have jobs and still be in their early careers, therefore,
#they'd have more free time.
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_sp = purchase_data.groupby(["SN", "Age"]).agg({"Price": ["count", "mean", "sum"]}).round(decimals=2)
top_sp_df = top_sp.sort_values([("Price", "sum")], ascending=False).head()
top_sp_df.rename(columns={'count': 'Purchase Count', 'mean': 'Average Price', 'sum': 'Total Price'})

#Observable trend 3

#All of the top spenders are in their 20s. This fits what was discovered above.
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = purchase_data.groupby(["Item ID", "Item Name"]).agg({"Price": ["count", "sum", "mean"]}).round(decimals=2)
items_df = items.sort_values([("Price", "count")], ascending=False).head()
items_df.rename(columns={'count': 'Purchase Count', 'mean': 'Average Price', 'sum': 'Total Price'})
```
